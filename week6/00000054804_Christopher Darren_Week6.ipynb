```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
superstore = pd.read_csv(r"D:\SEMESTER3\DATAANAYSIS\lab\week6\superstore_sales.csv")
superstore.head()
```

```python
superstore.info()
```

```python
superstore.shape
```

```python
#Change datatype

superstore["Order Date"] = pd.to_datetime(superstore["Order Date"], infer_datetime_format=True)
superstore["Ship Date"] = pd.to_datetime(superstore["Ship Date"], infer_datetime_format=True)
superstore.info()
```

```python
# Check Presentage of Missing Value, wil be handlede in next week

superstore.isnull().sum() / superstore.shape[0] * 100
```

```python
#Check values counts, kasus ini dimana menghitung how many customers dengan nama yang sama

superstore["Customer Name"].value_counts()
```

```python
#mengecek ada berapa unique values
#will be handled in data preparation(normal encoding )

superstore.groupby('Ship Mode')['Row ID'].nunique()
```

```python
superstore.groupby('Category')['Row ID'].nunique()
```

```python
superstore.groupby('Segment')['Row ID'].nunique()
```

```python
#flash back to some visualization

#boxplot/countplot
sns.set(rc= {'figure.figsize':(20,8)})
sns.countplot(x='Segment', data=superstore, palette="pastel")
```

```python
#atau kita bisa visualisasikan dengan pie chart

superstore.groupby('Segment')['Row ID'].count().plot(kind='pie',
                                                   labels=superstore['Segment'],
                                                   autopct='%.2f')
```

```python
#KDE plot

sns.set(rc={'figure.figsize':(20,8)})
sns.kdeplot(x='Sales', hue='Segment', data=superstore,palette="pastel", bw_adjust=20)
```

```python
#these variable, only "sales" is relevant for skewness

superstore.skew(axis=0, skipna=True)
```

```python
#Distribution plot

sns.set(rc= {'figure.figsize':(20,8)})
sns.displot(x='Sales',
           hue='Ship Mode',
           col='Segment',
           data=superstore,
           kind="kde",
           palette="pastel",
           bw_adjust=20)
```

```python
#Box plot, used to find outlier

sns.boxplot(x='Ship Mode', y="Sales", data=superstore, palette="pastel")
```

```python
#Correlation dengan heatmap

#data ini masih capable untuk korelasi data
sns.heatmap(superstore.corr(), annot=True)
```
