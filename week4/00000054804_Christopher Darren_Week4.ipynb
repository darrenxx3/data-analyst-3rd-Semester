```python
vaccine = pd.read_csv(r"D:\SEMESTER3\DATAANAYSIS\lab\week4\vaccination_data.csv")
vaccine.head()
```

```python
covid = pd.read_csv(r"D:\SEMESTER3\DATAANAYSIS\lab\week4\covid_data.csv")
covid.head()
```

```python
#CONVERT the data VARIABLE to date datatypes

vaccine["DATE_UPDATED"] = pd.to_datetime(vaccine["DATE_UPDATED"])
vaccine.info()
```

```python
#CHANGE column name if needed,

vaccine.rename(columns = {'PERSONS_VACCINATED_1PLUS_DOSE':'DOSE_1',
                          'PERSONS_FULLY_VACCINATED':'DOSE_2',
                          'PERSONS_BOOSTER_ADD_DOSE':'BOOSTER'
                         }, inplace = True)
vaccine.info()
```

```python
#SELECT part of variable from DB
#SELECT country, total vaccination 

#CREATE NEW DB, using df-> dataframe
df = vaccine[['COUNTRY','TOTAL_VACCINATIONS','DOSE_1','DOSE_2','BOOSTER']]
df.head()
```

```python
#CREATE one column to add category of total vaccinations

import numpy as np
df['COUNT_CATEGORY'] = np.where(
    df['TOTAL_VACCINATIONS'] > 5000000, 'Huge', np.where(
    df['TOTAL_VACCINATIONS'] < 5000000, 'Small', -1))
df.head()
```

```python
#CREATE new DF with dropped column

df_dropped = df.drop('COUNT_CATEGORY', axis=1)
df_dropped.head()
```

```python
#FILTERING COVID DATA, select only Indonesia

covid[covid['Country']== 'Indonesia']
```

```python
#FILTERING covid Data, select Indonesia and republic Korea

covid[covid['Country'].isin(['Indonesia','Republic of Korea'])]
```

```python
#FILTERING using part of the country keyword "Ko"

covid[covid['Country'].str.contains('Ko')]
```

```python
#Sometimes we need to pivot the data

#Convert the Data variable to datatypes
covid['Date_reported'] = pd.to_datetime(covid['Date_reported'])

#Transform date to Quarter
covid['Quarter'] = covid['Date_reported'].dt.to_period('Q')

#Pivoting
covid_pivot = covid.pivot_table(index='Country',
                                columns='Quarter',
                                values='New_cases',
                                aggfunc='sum')
covid_pivot.head(10)
```

```python
#GROUP BY FUNCTION

covid_group1 = covid.groupby(by="Country").sum() #JUST one Group and SUM all data
covid_group1.head()
```

```python
# IF we need some adjustment , we can use AGGREGATE function

covid_group2 = covid.groupby('Country').agg({
    'New_cases':['mean','max'],
    'New_deaths':'sum'})
covid_group2.head()
```

```python
#We will use aggregate to gather sum of new cases and deaths

covid_group = covid.groupby('Country').agg({
    'New_cases':'sum',
    'New_deaths':'sum'}).reset_index()
covid_group.head()
```

```python
#Join Data

covid_vaccine = covid_group.merge(vaccine,
                                 left_on='Country',
                                 right_on='COUNTRY',
                                 suffixes=(False, False),
                                 how='inner')
covid_vaccine
```

```python
#a.	What is the top 5 country that has the most death ratio?
df = covid_vaccine[['COUNTRY','New_deaths','New_cases']]
df.head()

df['DeathRatio'] = df['New_deaths'] / df['New_cases']
df
```
