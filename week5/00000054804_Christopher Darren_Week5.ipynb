```python
nba = pd.read_csv(r"D:\SEMESTER3\DATAANAYSIS\lab\week5\nba.csv")
nba.head()
```

```python
nba.info()
```

```python
nba.describe()
```

```python
sns.heatmap(nba.isnull(),cbar=False, cmap='YlGnBu_r')
sns.set(rc = {'figure.figsize':(20,8)})
plt.title('missing data')
plt.show()
```

```python
#Box plot

colors = sns.color_palette('pastel')[0:5]
#mendefinisikan
sns.countplot(x='Age', data=nba, palette=colors)

#mendefinisikan judul
plt.title('Age Distribution')
plt.show()
```

```python
#line kde plot

sns.kdeplot(x='Age', hue='Position', data=nba)
plt.title('Age on Position')
plt.show()
```

```python
#bar plot

colors = sns.color_palette('pastel')[3:4]
sns.histplot(data=nba, x='Height', color=colors)
```

```python
#horizontal Bar plot

colors = sns.color_palette("Set2")[0:1]
sns.histplot(data=nba, y="Height", color=colors)
```

```python
# boxplot


sns.set_theme(style="whitegrid")
ax = sns.boxplot(x=nba["Salary"])
```

```python
#box plot data vertical

colors = sns.color_palette('pastel')
sns.boxplot(x="Position", y="Salary", data=nba, palette=colors)
```

```python
#boxplot example, add variable "Hue"

colors = sns.color_palette('pastel')
sns.boxplot(x="Team", y="Salary", hue="Position", data=nba, palette=colors)
```

```python
#Distribusi plot

sns.displot(nba, x="Weight", hue="Position", kind="kde")
```

```python
#distribusi plot dengan warna

sns.displot(nba, x="Weight", hue="Position", kind="kde", multiple="stack")
```

```python
#scatter plot

sns.scatterplot(data=nba, x="Weight", y="Salary", color="blue")
```

```python
#scatter plot

sns.scatterplot(data=nba, x="Weight", y="Salary", hue="Position", palette="pastel")
```

```python
#scatter plot

sns.scatterplot(data=nba, x='Weight', y="Age",
               hue="Position", size="Salary",
               sizes=(40,300), palette="pastel")
```

```python
#heat map will be furthe discused at EDA

sns.heatmap(nba[['Weight','Age','Salary']].corr(), annot=True)
```

```python
sns.pairplot(data=nba, dropna=True, hue="Salary")
```
